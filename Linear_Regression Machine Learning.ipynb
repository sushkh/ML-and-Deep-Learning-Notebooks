```python
# Drop "'CarName', 'symboling', 'fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 
#       'enginetype', 'cylindernumber', 'fuelsystem'" as we have created the dummies for it

df_car.drop(varlist, axis = 1, inplace = True)
```

```python
df_car.drop(['car_ID'], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(df_car, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
df_train.columns
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight',
       'enginesize', 'boreratio', 'stroke', 'compressionratio', 'horsepower',
       'peakrpm', 'citympg', 'highwaympg', 'price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train[['carwidth','enginesize','curbweight','horsepower']])

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Let's visualise the data with a scatter plot and the fitted regression line
plt.scatter(X_train_lm.iloc[:, 1], y_train)
plt.plot(X_train_lm.iloc[:, 1], 0.127 + 0.462*X_train_lm.iloc[:, 1], 'r')
plt.show()
```

```python
# Print a summary of the linear regression model obtained
print(lr.summary())
```

```python
# Calculate the VIFs for the new model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
